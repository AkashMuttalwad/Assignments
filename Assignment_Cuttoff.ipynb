```python
import pandas as pd
import pandas as pd
from datetime import date
```

```python
data = pd.read_csv ('/content/drive/MyDrive/Assignment_cuttoff1_assignment.csv')
data
```

```python
max_total=data.groupby(['Coustomer ID', 'Date']).agg(max=('Brand count', 'max'),total=('Brand count', 'sum')).reset_index()
max_total
```

```python
max_total['cutoff']= (max_total['max']/max_total['total'])*100
max_total
```

```python
min_total=data.groupby(['Coustomer ID', 'Date']).agg(min=('Brand count', 'min'),total=('Brand count', 'sum')).reset_index()
min_total
```

```python
min_total['cutoff']= (min_total['min']/min_total['total'])*100
min_total
```
